```python
csvfile = open('documents/AICOURSES.csv', 'w', encoding='utf-8')  # w = write into the file , general way of representing data
writer = csv.writer(csvfile)
writer.writerow(['company','title', 'rating', 'metadata'])

for i in range(84):
    if i >= 1:
        url = f"https://www.coursera.org/search?query=ai%20courses&page={i}&sortBy=BEST_MATCH&=null"

    website = requests.get(url).text
    website = BeautifulSoup(website, "lxml")

    courses = website.find_all("div", class_="cds-CommonCard-clickArea")

    for course in courses:
        company = course.find("div", class_="cds-ProductCard-partnerInfo").text
        rating = course.find("div", class_="cds-CommonCard-ratings").text
        metadata = course.find("div", class_="cds-CommonCard-metadata").text
        title=course.find("a", class_ ="cds-119 cds-113 cds-115 cds-CommonCard-titleLink css-si869u cds-142").text
        writer.writerow([company,title, rating, metadata])

csvfile.close()
```
